````python
### probe data is stored as a json file and we will have to extract the information using the json package. 
# We will be using the json package to work with some of our segmentation files later.

import json
import pandas as pd
import os
````

````python
data_dir = '/home/shamini/data1/data_orig/data/spatial/xenium/10xGenomics/cell_seg_brain_cancer/'
out = '/home/shamini/data/projects/spatial_workshop/out/module4/'
os.makedirs(out, exist_ok=True)

os.makedirs(out+'objects', exist_ok=True)
os.makedirs(out+'figures', exist_ok=True)
````

````python

# Open JSON file
f = open(data_dir+'/Xenium_V1_Human_Brain_GBM_FFPE_gene_panel.json') # Edit file name here

# Return JSON object as a dictionary
data = json.load(f)

# Create lists to store extracted information
gene = []
ensembl = []
cov = []
# Iterate through the JSON list to extract information
for i in data['payload']['targets']:
    if (i['type']['descriptor'] == "gene"): # Only collect info for genes, not controls
        gene_name = i['type']['data']['name']
        ensembl_id = i['type']['data']['id']
        coverage = str(i['info']['gene_coverage'])

        gene.append(gene_name)
        ensembl.append(ensembl_id)
        cov.append(coverage)

# Create output CSV file
out_df = pd.DataFrame(list(zip(gene, ensembl, cov)), columns=['Gene name', 'Ensembl ID', 'Gene coverage'])
os.makedirs(data_dir+'module4', exist_ok=True)
out_df.to_csv(data_dir+'module4/panel_gene_info.csv', index=False)

# Close file
f.close()
````
